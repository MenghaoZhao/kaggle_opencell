```python
def save_submit(submit, version, i):
    
    submit.columns = de_train.columns[5:]
    submit.index.name = 'id'
    
    submit.to_csv(f'submission_v{version}_i{i}.csv')
```

```python
from pandas import read_parquet
adata_sorted = read_parquet("./adata_sorted.parquet")
adata_meta = pd.read_csv("./adata_obs_meta.csv")
for donor_num in range(3):
    adata_meta[f"donor_{donor_num}"] = (adata_meta["donor_id"]==f"donor_{donor_num}").astype(int)
adata_meta
```

```python
import numpy as np
# np.save("gene_idx.npy",np.array(gene_idx))
gene_idx = np.array(np.load("gene_idx.npy",allow_pickle=True))
gene_idx_dict = dict()
for ii in range(len(gene_idx)):
    gene_idx_dict[gene_idx[ii][0]] = [int(gene_idx[ii][1]), int(gene_idx[ii][2])]
```

```python
from tqdm import tqdm

from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import LinearSVR


N_NEIGHBORS = [20, 25]
MAX_ITER = [5000, 7000]
ID_MAP_LEN = len(id_map)
VERSION = 15
I_TOTAL = 99
i_total = I_TOTAL
i_blend_submit = 0

Y_submit_preds = np.zeros((ID_MAP_LEN,  Y.shape[1]))

for n_neighbors in [0]:#N_NEIGHBORS:
    
    for max_iter in MAX_ITER:
        
        i_total += 1
        model_knr = KNeighborsRegressor(n_neighbors=n_neighbors)
        model_lsvr = LinearSVR(max_iter=max_iter, epsilon=0.1)
        
        Y_preds = []
        
        for gene in tqdm(range(Y.shape[1])):

            gene_name = de_train.columns[gene+5]
            adata_obs = adata_sorted.iloc[gene_idx_dict[gene_name][0]:gene_idx_dict[gene_name][1]].merge(adata_meta,on="obs_id")
            adata_cell_mean = adata_obs[["cell_type","dose_uM","count","normalized_count"]].groupby("cell_type").agg("mean")
            adata_sm_mean = adata_obs[["sm_name","dose_uM","count","normalized_count"]].groupby("sm_name").agg("mean")
            adata_cell_count = adata_obs[["cell_type","dose_uM"]].groupby("cell_type").count().rename(columns={"dose_uM":"num_obs"})
            adata_sm_count = adata_obs[["sm_name","dose_uM"]].groupby("sm_name").count().rename(columns={"dose_uM":"num_obs"})
            adata_cell_donor = adata_obs[["cell_type","donor_0","donor_1","donor_2"]].groupby("cell_type").agg("sum")
            adata_sm_donor = adata_obs[["sm_name","donor_0","donor_1","donor_2"]].groupby("sm_name").agg("sum")

            train_X = de_train[["cell_type","sm_name",gene_name]]
            cell_mean = train_X.rename(columns={gene_name:gene_name+"_cell_type"}).groupby("cell_type")[gene_name+"_cell_type"].agg("mean")#
            sm_mean = train_X.rename(columns={gene_name:gene_name+"_sm_name"}).groupby("sm_name")[gene_name+"_sm_name"].agg("mean")#.rename(columns={gene_name:gene_name+"_sm_name"})

            train_X = train_X.merge(adata_cell_mean,on="cell_type",how="left")
            train_X = train_X.merge(adata_sm_mean,on="sm_name",how="left",suffixes=("_cell_type","_sm_name"))
            train_X = train_X.merge(adata_cell_count,on="cell_type",how="left")
            train_X = train_X.merge(adata_sm_count,on="sm_name",how="left",suffixes=("_cell_type","_sm_name"))
            train_X = train_X.merge(adata_cell_donor,on="cell_type",how="left")
            train_X = train_X.merge(adata_sm_donor,on="sm_name",how="left",suffixes=("_cell_type","_sm_name"))
            train_X = train_X.merge(cell_mean,on="cell_type",how="left")
            train_X = train_X.merge(sm_mean,on="sm_name",how="left",suffixes =("_cell_type","_sm_name"))

            train_X = train_X.drop(columns=["cell_type","sm_name",gene_name]).fillna(0)
            
            test_X = id_map[['cell_type','sm_name']]
            test_X = test_X.merge(adata_cell_mean,on="cell_type",how="left")
            test_X = test_X.merge(adata_sm_mean,on="sm_name",how="left",suffixes=("_cell_type","_sm_name"))
            test_X = test_X.merge(adata_cell_count,on="cell_type",how="left")
            test_X = test_X.merge(adata_sm_count,on="sm_name",how="left",suffixes=("_cell_type","_sm_name"))
            test_X = test_X.merge(adata_cell_donor,on="cell_type",how="left")
            test_X = test_X.merge(adata_sm_donor,on="sm_name",how="left",suffixes=("_cell_type","_sm_name"))
            test_X = test_X.merge(cell_mean,on="cell_type",how="left")
            test_X = test_X.merge(sm_mean,on="sm_name",how="left",suffixes =("_cell_type","_sm_name"))

            test_X = test_X.drop(columns=["cell_type","sm_name"]).fillna(0)

            Y_gene = Y.iloc[:, gene].copy()

            X_gene = X.join(cell_type_tr.iloc[:, gene + 1]).copy()
            # X_gene = X_gene.join(sm_name_tr.iloc[:, gene + 1], lsuffix='_cell_type', rsuffix='_sm_name')
            X_gene = X_gene.join(train_X)
            test_gene = dummies_test.join(cell_type_te.iloc[:, gene + 1]).copy()
            # test_gene = test_gene.join(sm_name_te.iloc[:, gene + 1], lsuffix='_cell_type', rsuffix='_sm_name')
            test_gene = test_gene.join(test_X)
            # model_knr.fit(X_gene, Y_gene)
            model_lsvr.fit(X_gene, Y_gene)

            # Y_pred_knr = model_knr.predict(test_gene)
            Y_pred_lsvr = model_lsvr.predict(test_gene)

            Y_preds.append( Y_pred_lsvr )
        print(f'I_TOTAL: {i_total}\nI_BLEND_SUBMIT: {i_blend_submit}\nN_NEIGHBORS: {n_neighbors}\nMAX_ITER: {max_iter}')
        print(30 * '-')
        submit = pd.DataFrame(Y_preds).T
        if i_total != I_TOTAL + 1:
            save_submit(submit, VERSION, i_total) 
            
        Y_submit_preds = (Y_submit_preds * i_blend_submit + submit) / (i_blend_submit + 1)
        save_submit(Y_submit_preds, VERSION, i_blend_submit)    
        
        i_blend_submit += 1
```
